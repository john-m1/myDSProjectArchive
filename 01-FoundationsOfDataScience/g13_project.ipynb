# Overview

The Effects of Population Growth on Water Pollution in Ontario

The Ontario Ministry of Environment, Conservation and Parks maintains serveral monitoring stations throughout the province and monitors pollutants such as Lead, Phosphorus and monitors for akaline and pH levels. The Ontraio program is called 'The Provincial (Stream) Water Quality Monitoring Network (PWQMN)'. [Ref-1](#ref1).

Statitics Canada maintains census information of population centers.

We wish to determine if there is a relationship between population growth and the amount of polutants in the water.

```python
summary={}

for p in df['PARM'].unique():
    # print(p)
    df_t = df[df['PARM'] == p][["STATION","PARM","YEAR","RESULT","UNITS","ANALMETHOD","GEO_CODE","GEO_NAME"]].dropna()
    regions = df_t["GEO_NAME"].unique()
    # print(regions)
    df_t_group = df_t.groupby(["GEO_NAME","GEO_CODE","YEAR","UNITS"]).agg({"RESULT": "mean"})
    
    rangesum={}
    for r in regions:
        rangesum[r]={'length': len(df_t_group.loc[[r]]), 
                     'minyear': df_t_group.loc[[r]].index.get_level_values(2).to_list()[0],
                     'maxyear': df_t_group.loc[[r]].index.get_level_values(2).to_list()[-1]
                    }
        
    summary[p] = rangesum

    
```

```python
dfp={}
for p in summary:
  dfp[p]=pd.DataFrame.from_dict(summary[p],orient='index')

print(len(dfp))
print()

for p in dfp.keys():
  print(p)
  print(dfp[p][dfp[p].length > 19])
```
